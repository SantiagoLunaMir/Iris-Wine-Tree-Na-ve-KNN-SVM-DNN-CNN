## Decission Tree
A decision tree is non-parametric supervised learning algorithm, which is utilized for classification and regression task. It has tree structure, hirearchical, which means that has node, braches, internal nodes and leaf nodes.

Image from IBM:
image.png
The most interesting Algorithms for DEsicion Tree are the next:
1. ID3: This algorithm leverages entropy and information gain as metrics to evalute candidate splits.
2. C4.5: Is developed by Quinlan too. It can use information gain or gain ratios to evaluate split points within the decision trees.
3. CART: "Classification and regression Trees", introduced by Leo Breiman. This algorithm typically utilizes Dini impurity to identify the ideal attribute to split on. Measuring how often a randmly chosen attribute is misclassified. When evaluating using Gini impurity, a lower value is more ideal.
How do we get retroalimention to use the best option?
We use entropy and Information Gain.
Entropy is a cocpt more use in stems from information theory, which measures the impurity of the same values. It is defined by:

$$
\begin{equation}
Entropy(S) = -\sum_{c‚ààC} (c)log_2p(c)
\end{equation}
$$
Where:
- S represents the data set that entropy is calculated
- c represents the classes in set, S
- p(c) represents the proportion of data points that belong to class c to the number of total data points in set, S

Usually the entropy values can fall between 0 and 1. If all samples in data set, S, belong to one class, the entropy will be equal to 0. If half of the samples are classified as one class and the other half are in another class, entropy will be at its highest at 1. In order to select the best feature to split on and find the optimal decision tree, the attribute with the smallest amount of entropy should be used. Information gain represents the difference in entropy before and after a split on a given attribute. The attribute with the highest information gain will produce the best split as it is doing the best job at classifying the training data according to its target classification.

The principal point that must be consider are the nest ones:
- Advantages
  * Easy to interpret
  * Little to no data preparation requiered
  * More flexible
- Disadvantages
  * Prone to overfiting
  * High variance estimators
  * More costly


## Na√Øve Bayes
Is a supervised machine learning algorithm, which is used for classifiaction task, like text classification. It is part too for generative algorithms, this means that the model seeks to model the distribution of inputs of a given class or category but the algorithm does not learn which features are most important to differiantiate between classes. The algorithm is also know as a probabilistic classifier since it is based on Baye's Theorem. The toerem (that is known as Baye's Rule also), allows us to "invert" conditional probabilities. THe conditional probabilities represent the probability of and event given us information that some other event has already occured and is represented in the next formula:

$$
\begin{equation}
P(Y|X) = P(X and Y)/ P(x)
\end{equation}
$$

The theorem is distinguished for using sequuentians event, that means that the additional information later acquired impacts the initial probability. The prior probability is an initial probability of a certain event before it is contextualized under a certain condition, or the margianl probability. The probability that is posterior is the ones of an event after observing a piece of data. The theorem assumes that predictiors are conditionally independent, or underelated to any of the other feature in the model. It also assumes that all features contribute equally to the outcome.
One of the differnte ways to evaluate the model is to plot a confusion matrix, which will plot the actual and predicted values within a matrix. Rows usually represent actual values and while columns the predicted values.

## KNN algorithm
Is an algorithm which is non-parametric, supervised leaning classifier, which is uses proximity to make classifications or predictions from a group or individual points of data. It is one of the most popular and simplest classification and regression classifiers used in machine learning.
Is used typically as a classification algorithm, working off the assumption that siminlar ponts can be found near one another.
The algorithm use the "plurality voting" that means that is not necesary the "majority vote" (that means that requieres >50% of the 'votes') meanwhile the plurality voting says that take the resolt as the maximum value in the group for example from [25%-A,10%-B,5%-C,5%-D,3%-E,2%-F,20%-G,10%-H,20%-I]. The value that the algorithm consider to classify the group as an 'A' is the 25% cause it is the maximum value from the group.

The main distinction here is that classification is used for discrete values, whereas regression is used with continuous ones. However, before a classification can be made, the distance must be defined. Euclidean distance is most commonly used, which we‚Äôll delve into more below.
It's also worth noting that the KNN algorithm is also part of a family of ‚Äúlazy learning‚Äù models, meaning that it only stores a training dataset versus undergoing a training stage. This also means that all the computation occurs when a classification or prediction is being made. Since it heavily relies on memory to store all its training data, it is also referred to as an instance-based or memory-based learning method.
What are the metrics that we need to consider?
- Determinate distance metrics
Determinate which data points are the closest to a given query point, the distance between them will calculate. These distance metrics help to form decision boundaries. The most commonly distance measure are
1. Euclidean distance (p=2). Measured be the next formula:

$$
\begin{equation}
d(x,y) = \sqrt {\sum_{i=1}^{n}(yi-xi)^2}
\end{equation}
$$

2. Manhattan distance (p=1). Measured be the next formula:

$$
\begin{equation}
d(x,y) = (\sum_{i=1}^{m}|Xi-Yi|)
\end{equation}
$$

3. Minkowski distance (p=1). Measured be the next formula:

$$
\begin{equation}
Minkowski distance = (\sum_{i=1}^{n}|Xi-Yi|)^(1/p)
\end{equation}
$$

4. Hammming distance (p=1). Measured be the next formula:

$$
\begin{equation}
DH = (\sum_{i=1}^{k}|Xi-Yi|)
\end{equation}
$$

Where:
x=y   D=0
x!=y  D!=1


The value K in k-NN determinates how many neighborgs will be checked to determine the classification of a specific query point. For example, if k=1, the instance will be assigned to the same class as its single nearest neighbor. Defining k can be a balancing act as different values can lead to overfitting or underfitting. Lower values of k can have high variance, but low bias, and larger values of k may lead to high bias and lower variance. The choice of k will largely depend on the input data as data with more outliers or noise will likely perform better with higher values of k.

## Support Vector Machine (SVM)
Is a supervising learning algorithm used in machine learning to solve classification and regression tasks. This type of algorithms are particularly good at solving binary classification problems, where is requiered classifying the elements of data set into two groups.
The aim of support vector machine algorithm is to find the best possible line that separates the data points of different data classes. THis is called as hyperplane when working in high-dimensional feature spaces. The principal idea is to maximize the margin, which is the distance between the hyperplane and the closest data points of each category, which helps easy to distinguish data classes.
SVMs are useful for analyzing complex data that can't be separated by a simple straight line. Whit nonlinenar SVMs, they do this by using mathematicals that transform data into higher-dimensional space which makes easier find boundary.
The SVMs uses a kernel function to avoid the calcutation of each coordinate from the trasnformed space, the evalueates vectors and avoid handling expensive computations.
Sometimes an linear separable line is not the best option so we could use differente types of kernel functions, linear kernels, polynomials kernel or radial basis fuction (RBF) kernel. These kernels enable SVMs to effectively captura complex realtionships and pattarns in the data.
The best kernel function for a a particular problem depends on the characteristics of the data.

```python
# Load the wine dataset
wine = datasets.load_wine()
X = wine.data
y = wine.target.reshape(-1, 1)

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Encode labels into one-hot encoding
encoder = OneHotEncoder(sparse=False)
y_onehot = encoder.fit_transform(y)

# Split the dataset into training and testing (50% each)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_onehot, test_size=0.5, random_state=42)

# Create the DNN model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X.shape[1],)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(len(wine.target_names), activation='softmax')  # Number of classes in the wine dataset, the classes are the number of exits
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=16, validation_data=(X_test, y_test))

# Evaluate the model on the test set
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Loss on test set: {loss}')
print(f'Accuracy on test set: {accuracy}')
```
